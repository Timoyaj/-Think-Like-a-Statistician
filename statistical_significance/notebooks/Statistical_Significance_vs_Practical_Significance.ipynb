## Statistical Significance vs Practical Significance

This notebook demonstrates why statistical significance and practical significance are not the same.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

np.random.seed(42)
n=50000
old=np.random.normal(100,15,n)
new=np.random.normal(100.3,15,n)

t,p=stats.ttest_ind(new,old)
d=(new.mean()-old.mean())/np.sqrt((old.var(ddof=1)+new.var(ddof=1))/2)
print(f'p-value={p:.3e}')
print(f"Cohen's d={d:.3f}")

plt.hist(old,bins=40,alpha=.5,label='Old')
plt.hist(new,bins=40,alpha=.5,label='New')
plt.legend()
plt.title('Tiny Effect, Huge Sample')
plt.show()
```

## Interpretation

A very small p-value indicates evidence against the null hypothesis, but the effect size is tiny. Statistical significance does not automatically imply practical importance.

### Recommended reporting
- p-value
- Effect size (Cohen's d)
- Confidence interval
- Practical implications
